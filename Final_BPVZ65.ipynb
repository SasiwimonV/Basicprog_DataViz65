* คะแนน 60% จากกรรมการ(30%+30%)
* คะแนน 40% จากอาจารย์ที่ปรึกษา
1. ตัวอย่าง
* A ได้คะแนนจากกรรมการคนที่ 1  52/70
* A ได้คะแนนจากกรรมการคนที่ 2  60/70
* A ได้คะแนนจากกรรมการ = 52+60/140 คือ 80 หรือ 80%
* คะแนน 60% จากกรรมการของAทั้งสองคน = 60% ของ 80 คือ 48 คะแนน
* A ได้คะแนนจากอาจารย์ที่ปรึกษา  62/70 คือ 88.6 หรือ 88.6%
* คะแนน 40% จากอาจารย์ที่ปรึกษาของA = 40% ของ 88.6 คือ 35.44 คะแนน
* A จะได้คะแนนทั้งหมด 48+35.44 = 83.44 คะแนน

# 1.รวมคะแนนและตัดเกรดแบบ อิงเกณฑ์ <50 ได้ F ... >= 80 ได้ A และ save ตาราง ชื่อ คะแนนรวมและเกรดที่ได้ของนักศึกษาแต่ละคน
* ถ้าคะแนนมากกว่าหรือเท่ากับ 80 จะได้เกรด A
* ถ้าคะแนนอยู่ระหว่างมากกว่าหรือเท่ากับ 75 แต่น้อยกว่า 80 จะได้เกรด B+
* ถ้าคะแนนอยู่ระหว่างมากกว่าหรือเท่ากับ 70 แต่น้อยกว่า 75 จะได้เกรด B
* ถ้าคะแนนอยู่ระหว่างมากกว่าหรือเท่ากับ 65 แต่น้อยกว่า 70 จะได้เกรด C+
* ถ้าคะแนนอยู่ระหว่างมากกว่าหรือเท่ากับ 60 แต่น้อยกว่า 65 จะได้เกรด C
* ถ้าคะแนนอยู่ระหว่างมากกว่าหรือเท่ากับ 55 แต่น้อยกว่า 60 จะได้เกรด D+
* ถ้าคะแนนอยู่ระหว่างมากกว่าหรือเท่ากับ 50 แต่น้อยกว่า 55 จะได้เกรด D
* ถ้าคะแนนอยู่ระหว่างมากกว่าหรือเท่ากับ 0 แต่น้อยกว่า 50 จะได้เกรด F
