#ปริมาณฝน

```python
import pandas as pd
ksn = pd.read_excel('/content/drive/MyDrive/ข้อมูลกรมอุตุ/ปริมาณน้ำฝน/ปริมาณฝนกาฬสินธุ์.xlsx',header=[0])
kkc = pd.read_excel('/content/drive/MyDrive/ข้อมูลกรมอุตุ/ปริมาณน้ำฝน/ปริมาณฝนขอนแก่น.xlsx',header=[0])
cpm= pd.read_excel('/content/drive/MyDrive/ข้อมูลกรมอุตุ/ปริมาณน้ำฝน/ปริมาณฝนชัยภูมิ.xlsx',header=[0])
mkm = pd.read_excel('/content/drive/MyDrive/ข้อมูลกรมอุตุ/ปริมาณน้ำฝน/ปริมาณฝนมหาสารคาม.xlsx',header=[0])
nk = pd.read_excel('/content/drive/MyDrive/ข้อมูลกรมอุตุ/ปริมาณน้ำฝน/ปริมาณฝนหนองคาย.xlsx',header=[0])

ksn['จังหวัด'] = 'กาฬสินธุ์' #เพิ่มคอลัมน์จังหวัด
kkc['จังหวัด'] = 'ขอนแก่น'
cpm['จังหวัด'] = 'ชัยภูมิ'
mkm['จังหวัด'] = 'มหาสารคาม'
nk['จังหวัด'] = 'หนองคาย'
```

```python
ksn.columns
```

```python
Speed = pd.concat([ksn,kkc,cpm,mkm,nk])
Speed.shape
```

```python
Speed.columns
```

```python
Speed['เดือน'] = Speed['เดือน/ปี'].dt.month
Speed['ปี'] = Speed['เดือน/ปี'].dt.year
del Speed['เดือน/ปี']
Speed.head()
```

```python
Speed = Speed[['ที่', 'สถานี', 'รวม','จังหวัด']]
```

```python
Speed.head()
```
